```python
# Install necessary libraries
!pip install -q transformers datasets torchaudio accelerate

# Imports
import torch
import torchaudio
from transformers import WhisperProcessor, WhisperForConditionalGeneration

# Load model and processor
processor = WhisperProcessor.from_pretrained("openai/whisper-medium")
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-medium")

# Set device
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
```

```python
def transcribe_kannada(audio_path):
    # Load audio
    audio, rate = torchaudio.load(audio_path)

    # Resample to 16000Hz if needed
    if rate != 16000:
        resampler = torchaudio.transforms.Resample(orig_freq=rate, new_freq=16000)
        audio = resampler(audio)

    # Tokenize
    inputs = processor(audio.squeeze(), sampling_rate=16000, return_tensors="pt")
    input_features = inputs.input_features.to(device)

    # Generate transcription
    generated_ids = model.generate(
        input_features,
        language="kannada",
        task="transcribe"
    )

    # Decode and return
    transcript = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
    return transcript
```

```python
import os
import csv

# Folder containing your audio files
folder_path = "/content/drive/MyDrive/INTERNSHIP/female"

# Output CSV path in Drive
csv_output_path = "/content/drive/MyDrive/INTERNSHIP/kannada_transcriptions.csv"

# Counter and list for saving rows
count = 0
transcriptions = []

# Loop through all .wav files in the folder
for filename in sorted(os.listdir(folder_path)):
    if filename.endswith(".wav"):
        audio_path = os.path.join(folder_path, filename)
        try:
            transcript = transcribe_kannada(audio_path)
            count += 1
            print(f"{count:03d}. 📝 {filename} Transcript: {transcript}")
            transcriptions.append([count, filename, transcript])
        except Exception as e:
            print(f"❌ Error processing {filename}: {e}")
            count += 1
            transcriptions.append([count, filename, f"Error: {e}"])

        # Stop after 100 files
        if count >= 100:
            break

# Save results to CSV
with open(csv_output_path, mode='w', encoding='utf-8', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["S.No", "Filename", "Transcript"])
    writer.writerows(transcriptions)

print(f"\n✅ Transcriptions saved to: {csv_output_path}")
```

```python
import csv
import os
from jiwer import wer, cer

# === File Paths ===
csv_pred_path = "/content/drive/MyDrive/INTERNSHIP/kannada_transcriptions.csv"
tsv_ref_path = "/content/drive/MyDrive/INTERNSHIP/line_index_female.tsv"
output_path = "/content/drive/MyDrive/INTERNSHIP/evaluation_results.csv"

# === Load reference TSV into dictionary ===
ref_dict = {}
with open(tsv_ref_path, 'r', encoding='utf-8') as f:
    for line in f:
        parts = line.strip().split('\t')
        if len(parts) >= 2:
            filename = os.path.splitext(parts[0].strip())[0]  # Remove extension
            reference = parts[1].strip()
            ref_dict[filename] = reference

# === Process predictions and compute scores ===
results = []
all_refs = []
all_preds = []

with open(csv_pred_path, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        s_no = row.get("S.No", "").strip()
        filename_raw = row.get("Filename", "").strip()
        predicted = row.get("Transcript", "").strip()

        # Normalize filename (remove path + extension)
        filename = os.path.splitext(os.path.basename(filename_raw))[0]

        reference = ref_dict.get(filename, "").strip()

        if reference:
            wer_score = round(wer(reference, predicted), 4)
            cer_score = round(cer(reference, predicted), 4)
            all_refs.append(reference)
            all_preds.append(predicted)
        else:
            wer_score = cer_score = "N/A"

        results.append([s_no, filename_raw, reference, predicted, wer_score, cer_score])

# === Save per-file evaluation results ===
with open(output_path, 'w', encoding='utf-8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(["S.No", "Filename", "Reference", "Predicted", "WER", "CER"])
    writer.writerows(results)

# === Compute overall WER & CER ===
if all_refs:
    overall_wer = round(wer(all_refs, all_preds), 4)
    overall_cer = round(cer(all_refs, all_preds), 4)
    print(f"✅ Evaluation results saved at: {output_path}")
    print(f"📊 Overall WER: {overall_wer}")
    print(f"📊 Overall CER: {overall_cer}")
else:
    print("⚠️ No valid references matched the predictions. Check filename formats.")
```
